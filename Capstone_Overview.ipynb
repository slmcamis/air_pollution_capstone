# Air Pollution in the United States

Data obtained from https://aqs.epa.gov/aqsweb/airdata/download_files.html

The EPA collects data on 6 main pollutants: ozone ($O_{3}$), sulphur dioxide ($SO_{2}$), carbon monoxide ($CO$), nitrogen dioxide ($NO_{2}$), particulate matter ($PM_{10}$), and fine particulate matter ($PM_{2.5}$). They also track daily meterological data including temperature, winds, and barometric pressure. 

$PM_{10}$ is particulate matter 10 micrometers or less in diameter, $PM_{2.5}$ is particulate matter 2.5 micrometers or less in diameter. $PM_{2.5}$ is generally described as fine particles. For comparison, a human hair is about 100 micrometres in diameter.

Air quality can be affected by the presence of certain pollutants. The main pollutants in this database are:
* Carbon monoxide (CO)
* Nitrogen dioxide (NO2)
* Particulate matter (PM)
* Ozone (O3)
* Sulphur dioxide (SO2)

Air quality is measured by the Air Quality Index (AQI). The AQI is measured with respect to each different pollutant in discrete time intervals.
